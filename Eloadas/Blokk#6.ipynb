## <center>Felező módszer, Húrmódszer és Newton-módszer</center>

<br>
<b>Példa.</b> Tekintsük az

$$f(x)=e^x-4x^2=0,\quad x\in [3,5]$$

egyenletet. Határozzuk meg az egyenlet gyökét a felező módszer segítségével a tanult leállási feltételek esetén!


```octave
% Intervallumfelezes
clear all
a=3; b=5; fuggveny='exp(x)-4*x^2'; maxit=100; TOL=10^-6;

format long
abszhiba = inf;
k=0;
while k<maxit && abszhiba>TOL
   k=k+1;
   x=a+(b-a)/2;
      fx=eval(fuggveny);
   if fx>0 
      b=x;
   else
      a=x;
   end;
   abszhiba = abs(b-a)
end;
k
x
```

```octave
x = 4.2:0.01:4.4;
plot(x,exp(x)-4*x.^2)
```

```octave
format long
% function [iteracioszam x_numerikus_ertek] = hurmodszer(fuggveny,a,b,maxit,TOL)
[iteracioszam x_numerikus_ertek] = hurmodszer('exp(x)-4*x^2',3,5,100,1e-6)
```

```octave
% [iteracioszam x_numerikus_ertek] = newtonmodszer(fuggveny,derivalt,a,b,x,maxit,TOL)
[iteracioszam x_numerikus_ertek] = newtonmodszer('exp(x)-4*x^2','exp(x)-8*x',3,5,4,100,1e-6)
```
