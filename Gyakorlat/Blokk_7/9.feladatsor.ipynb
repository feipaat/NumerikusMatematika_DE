#### <center>Kétváltozós szélsőérték keresés</center>

<br>
<b>1.feladat.</b> Tekintsük az alábbi 

$$f(x_1,x_2)=\frac{1}{4}\left(2x_1^3+3x_1^2-12x_1\right)+\frac{1}{2}\left(2x_2^3+3x_2^2-12x_2\right)$$

kétváltozós függvényt. A beépített <i>meshgrid</i>, <i>surf</i> és <i>contourf</i> parancsok segítségével ábrázoljuk a kétváltozós függvényt és annak szintvonalait a $[-3, 2]\times [-3, 2]$ négyzeten!

```octave
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;

f=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[X_1, X_2] = meshgrid(x_1, x_2);
surf(X_1, X_2, f(X_1,X_2));
figure;
contourf(X_1, X_2, f(X_1,X_2));
```

<br>
<b>2.feladat.</b> Az előző feladat szintvonalához tartozó ábrára tegyük rá a megfelelő szintvonalértékeket, illetve ábrázoljuk a pozitív, majd negatív gradiens mezőket is! Ehhez használjuk a beépített <i>contour</i>, <i>clabel</i>, <i>gradient</i> és <i>quiver</i> parancsokat!



```octave
[C,h] = contour(X_1,X_2,f(X_1,X_2));
clabel(C,h,'FontSize',12,'Color','black')
```

```octave
figure
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12,'Color','black')
hold on
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
quiver(x_1,x_2,DX,DY)
title('A fuggveny szintvonalai es a pozitiv gradiens mezo')
hold off
```

```octave
figure
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(-Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12,'Color','black')
hold on
[x_1,x_2] = meshgrid(-3:.6:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(-Z);
quiver(x_1,x_2,DX,DY)
title('A fuggveny szintvonalai es a negativ gradiens mezo')
hold off
```

<b>3.feladat.</b> Egészítsük ki az előző feladatban szereplő ábrát (szintvonalak értékeivel és pozitív gradiens mezővel) a kiszámolt $(-2,-2),\ (-2,-1),\ (1,-2),\ (1,1)$ stacinárius pontokkal!

```octave
x_1 = -3:0.05:2;
x_2 = -3:0.05:2;
x=@(x_1,x_2)((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
colormap("white")
[X_1, X_2] = meshgrid(x_1, x_2);
hold on
contourf(X_1, X_2, x(X_1,X_2));
plot(1,1,'m*',1,-2,'m*',-2,1,'m*',-2,-2,'m*')
[x_1,x_2] = meshgrid(-3:.4:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
quiver(x_1,x_2,DX,DY)
title('A fuggveny stacionarius pontjai, szintvonalai es a pozitiv gradiens mezo')
[x_1,x_2] = meshgrid(-3:.2:2);
Z = ((1/4)*(2*x_1.^3+3*x_1.^2-12*x_1)+(1/2)*(2*x_2.^3+3*x_2.^2-12*x_2));
[DX,DY] = gradient(Z);
[C,h] = contour(x_1,x_2,Z);
clabel(C,h,'FontSize',12)
hold off

```
